```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df=pd.read_csv("Obesity.csv")
```

```python
df.head()
```

```python
# Normalize numerical columns
scaler = MinMaxScaler()
numerical_columns = ['Age', 'Height', 'Weight', 'FCVC', 'CH2O', 'FAF', 'TUE']
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Encode categorical columns
encoder = LabelEncoder()
categorical_columns = ['Gender', 'family_history', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS', 'Obesity']
encoded_data = {}

for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])
    encoded_data[col] = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

# Save encoding mapping
encoded_data_df = pd.DataFrame.from_dict(encoded_data, orient='index').transpose()
encoded_data_df.to_csv('encoding_mapping.csv', index=False)
```

```python
# Print encoding
print("Encoding mappings:")

print(encoded_data)
```

```python
df['Binary'] = df['Obesity'].map(lambda x: 0 if x == 1 else 1)

df["Binary"].unique()
```
